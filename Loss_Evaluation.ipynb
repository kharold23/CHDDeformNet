```python
import pandas as pd
import numpy as np

dice_1 = pd.read_csv('/scratch/users/kharold/Validate/T1/ct_test.csv', skiprows=1, names=['overall', 'LV', 'RV', 'Myo'])

print('Results from new weights:\n')

print('Dice Score Results:')
dice_1_results = pd.DataFrame({' ': ['Mean', 'Std'],'overall': [np.mean(dice_1['overall']), np.std(dice_1['overall'])], 'LV': [np.mean(dice_1['LV']), np.std(dice_1['LV'])], 'RV': [np.mean(dice_1['RV']), np.std(dice_1['RV'])], 'Myo': [np.mean(dice_1['Myo']), np.std(dice_1['Myo'])]})
dice_1_results.loc[len(dice_1_results.index)] = [' ', ' ', ' ', ' ', ' ']
dice_1_results.loc[len(dice_1_results.index)] = ['Best Result'] + dice_1[dice_1['overall'] == np.max(dice_1['overall'])].values.flatten().tolist()
dice_1_results.loc[len(dice_1_results.index)] = ['Worst Result'] + dice_1[dice_1['overall'] == np.min(dice_1['overall'])].values.flatten().tolist()

print(dice_1_results.to_string(index=False))

assd_1 = pd.read_csv('/scratch/users/kharold/Validate/T1/ct_test_assd.csv', skiprows=1, names=['overall', 'LV', 'RV', 'Myo'])
assd_1 = assd_1.multiply(0.1)

print('\n\nAssd Results:')
assd_1_results = pd.DataFrame({' ': ['Mean', 'Std'],'overall': [np.mean(assd_1['overall']), np.std(assd_1['overall'])], 'LV': [np.mean(assd_1['LV']), np.std(assd_1['LV'])], 'RV': [np.mean(assd_1['RV']), np.std(assd_1['RV'])], 'Myo': [np.mean(assd_1['Myo']), np.std(assd_1['Myo'])]})
assd_1_results.loc[len(assd_1_results.index)] = [' ', ' ', ' ', ' ', ' ']
assd_1_results.loc[len(assd_1_results.index)] = ['Best Result'] + assd_1[assd_1['overall'] == np.min(assd_1['overall'])].values.flatten().tolist()
assd_1_results.loc[len(assd_1_results.index)] = ['Worst Result'] + assd_1[assd_1['overall'] == np.max(assd_1['overall'])].values.flatten().tolist()

print(assd_1_results.to_string(index=False))
```
